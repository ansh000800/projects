```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('project1.csv',encoding='latin-1')
# use the encoding code latin-1 for ignoring the unicode error.
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
# use drop function for droping the columns from the dataframe by axis 1 we delete the whole column and inplace true saves this command and make changes according to it keeps the dataframe according to command.
```

```python
pd.isnull(df).sum()
# finding the null values in dataframe.
```

```python
df.dropna(inplace=True)
# dropping the null values from data frame it delete the all row of null value.
```

Exploratory data analysis using seaborn

```python
# all columns name
df.columns
```

```python
# 1 for age group.
gender=sns.countplot(x='Gender',data=df)
# for placing the values on bars we use following trick in sns
for bars in gender.containers:
    gender.bar_label(bars)

# from above we can see that females are buying most.
```

```python
status=sns.countplot(x='Marital_Status',data=df,hue='Gender')
for bars in status.containers:
    status.bar_label(bars)

# from above we can see that non married are buying more.
```

```python
# now we are creating a bar chart for gender vs total amount.
gen_amount=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=gen_amount)
```

```python
# now for age group.
age_grp=sns.countplot(data=df,x="Age Group")
for bars in age_grp.containers:
    age_grp.bar_label(bars)
```

```python
# now diving the age group in male and females further.
age_gen=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in age_gen.containers:
    age_gen.bar_label(bars)
```

```python
# total amount and age group
age_amt=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=age_amt,x='Age Group',y='Amount')
```

```python
# now we see which top 5 states ordered the most.
state_ord=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,10)})
# here by sns.set we set the size of bar graph.
sns.barplot(data=state_ord,x='State',y='Orders')
```

```python
# now we see which top 5 states spends the more.
state_amt=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)
sns.set(rc={"figure.figsize":(15,10)})
sns.barplot(data=state_amt,x='State',y='Amount')
```

```python
# now we check that which occupation is shopping more.
sns.set(rc={'figure.figsize':(20,8)})
occ=sns.countplot(data=df,x='Occupation')
for bars in occ.containers:
    occ.bar_label(bars)
```

```python
# now check that which occupation sector spend the most.
occ_amt=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(data=occ_amt,x='Occupation',y='Amount')
```

```python
# now lets watch which category has placed the more orders.
sns.set(rc={'figure.figsize':(25,8)})
prd=sns.countplot(data=df,x='Product_Category')
for bars in prd.containers:
    prd.bar_label(bars)
```

```python
# now we are finding which category of prodct generates the mnore amount.
prd_amt=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=prd_amt,x='Product_Category',y='Amount')
```

```python
# now we will find which product id are the most selling .
prd_order=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(5)

sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(data=prd_order,x='Product_ID',y='Orders')
```
